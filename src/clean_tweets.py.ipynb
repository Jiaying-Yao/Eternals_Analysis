```python
import pandas as pd
import re
```

```python
tweets_df = pd.read_csv("../data/tweets.csv")
tweets_df = tweets_df.iloc[:,1:]
tweets_df
```

```python
pd.set_option('display.max_columns', None)
print(tweets_df['text'][1])
```

```python
clean = tweets_df
clean['text'] = clean['text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
clean
```

```python
clean = clean.drop_duplicates(subset=['text'])
clean = clean.drop_duplicates(subset=['author'])
clean
```

```python
clean.to_csv("../data/clean_tweets.tsv", index=False, sep='\t')
```

```python
extended_tw = pd.read_csv('../data/extended_clean_tw.tsv', sep = '\t')
extended_tw['full_text']= extended_tw['full_text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
extended_tw['full_text']
```

```python
extended_tw = extended_tw[extended_tw['full_text']!='']
extended_tw
```

```python
extended_tw.to_csv('../data/extended_clean_tweets.tsv', index = False, sep="\t")
```
